```python
import pandas as pd
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('./daily_weather.csv') 
```

```python
data.columns
```

```python
data.head()
```

```python
data[data.isnull().any(axis=1)]
```

```python
del data['number']
```

```python
data=data.dropna()
```

```python
clean_data=data.copy()
clean_data['high_humidity_label']=(clean_data['relative_humidity_3pm']>24.99) * 1 
print(clean_data['high_humidity_label'].head())
```

```python
y = clean_data[['high_humidity_label']].copy()
y.head()
```

```python
clean_data['high_humidity_label'].head()
```

```python
morning_feature = ['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am',
        'max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am',
        'rain_duration_9am']
```

```python
X = clean_data[morning_feature].copy()
```

```python
X.columns
```

```python
y.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
```

```python
X_train.head()
```

```python
type(X_test)
```

```python
y_train.describe()
```

```python
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
humidity_classifier.fit(X_train, y_train)
```

```python
type(humidity_classifier)
```

```python
predicitions = humidity_classifier.predict(X_test)
```

```python
predicitions[:10]
```

```python
y_test['high_humidity_label'][:10]
```

```python
accuracy_score(y_true = y_test, y_pred = predicitions)
```
